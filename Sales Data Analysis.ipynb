```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_excel('ECOMM DATA.xlsx')
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isna().sum()
```

```python
data.drop_duplicates(inplace=True)
```

Exploratory Data Analysis

```python
# Univariate Analysis: Histograms
data.hist(bins=50, figsize=(20, 15))
plt.show()
```

```python
# Aggregate sales by country
sales_by_country = data.groupby('Country')['Sales'].sum().reset_index()
sales_by_country=sales_by_country.sort_values(by='Sales',ascending=False).head(20)
```

```python
plt.figure()
plt.bar(sales_by_country['Country'], sales_by_country['Sales'])
plt.xticks(rotation=90)
plt.title('Top 20 countries having maximum sales')
plt.xlabel('Countries')
plt.ylabel('Revenue')
plt.ticklabel_format(style='plain',axis='y')
plt.show()
```

```python
# Aggregate profit by country
profit_by_country = data.groupby('Country')['Profit'].sum().reset_index()
profit_by_country=profit_by_country.sort_values(by='Profit',ascending=False).head(20)
```

```python
plt.figure()
plt.bar(profit_by_country['Country'], profit_by_country['Profit'])
plt.xticks(rotation=90)
plt.title('Top 20 countries with maximum profit')
plt.xlabel('Countries')
plt.ylabel('Profit')
plt.ticklabel_format(style='plain',axis='y')
plt.show()
```

```python
sns.countplot(x='Category',data=data)
plt.title("Categorywise count of products")
plt.show()
```

```python
# Define your preferred color palette
my_palette = sns.color_palette('Set1')

# Plot countplot for Category with Sub-Category as hue
plt.figure(figsize=(14, 8))
sns.set_palette(my_palette)  # Set the color palette
sns.countplot(x='Category', hue='Sub-Category', data=data)
plt.title('Count of Sub-Category within Each Category')
plt.xlabel('Category')
plt.ylabel('Count')
plt.legend(title='Sub-Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
data['Product ID'].nunique()
product_counts=data['Product ID'].value_counts()
product_counts.columns = ['Product ID', 'count']
product_counts
```

```python
# Get the top 15 product IDs with the highest counts
top_15_products = product_counts.head(15)

# Merge with original DataFrame to get sub-category information
top_15_products_with_subcategory = pd.merge(top_15_products, data, on='Product ID', how='left')

# Plotting using seaborn
plt.figure(figsize=(12, 8))
sns.barplot(x='Product ID', y='count', hue='Sub-Category', data=top_15_products_with_subcategory)
plt.title('Top 15 Product IDs with Counts and Sub-Category')
plt.xlabel('Product')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.legend(title='Sub-Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
data['Product Name'].nunique()
product_counts=data['Product Name'].value_counts()
product_counts.columns = ['Product Name', 'count']
product_counts
```

```python
# Get the top 15 products with the highest counts
top_15_products = product_counts.head(15)

# Merge with original DataFrame to get sub-category information
top_15_products_with_subcategory = pd.merge(top_15_products, data, on='Product Name', how='left')

# Plotting using seaborn
plt.figure(figsize=(12, 8))
sns.barplot(x='Product Name', y='count', data=top_15_products_with_subcategory)
plt.title('Top 15 Product names with Counts')
plt.xlabel('Product')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Aggregate quantity by sub-category
Quantity_by_subcat = data.groupby('Sub-Category')['Quantity'].sum().reset_index()
Quantity_by_subcat = Quantity_by_subcat.sort_values(by='Quantity',ascending=False).head(20)
Quantity_by_subcat
```

```python
# Create the horizontal bar plot
plt.figure(figsize=(10, 6))
plt.barh(Quantity_by_subcat['Sub-Category'], Quantity_by_subcat['Quantity'], color='skyblue')

# Adding titles and labels
plt.title('Sub-Category wise total order Quantity')
plt.xlabel('Quantity')
plt.ylabel('Sub-Category')
plt.show()
```

```python
# Create 'Month' and 'Year' columns
data['Month_Year'] = data['Order Date'].dt.strftime('%m-%Y')
data['Year'] = data['Order Date'].dt.year
data.head()
```

```python
# month-yearwise sales
Sales_by_month = data.groupby('Month_Year')['Sales'].sum().reset_index()
Sales_by_month = Sales_by_month.sort_values('Month_Year')
```

```python
Sales_by_month['Month_Year'] = pd.to_datetime(Sales_by_month['Month_Year'],format='%m-%Y')
Sales_by_month['Year'] = Sales_by_month['Month_Year'].dt.year
Sales_by_month['Month'] = Sales_by_month['Month_Year'].dt.month
years=data['Year'].unique()
plt.figure()

for year in years:
    data_year=Sales_by_month[Sales_by_month['Year']==year]
    plt.plot(data_year['Month_Year'],data_year['Sales'],marker='o',linestyle='-',label=f'Year {year}')
plt.title('Yearwise Monthly sales')
plt.xlabel('Month year')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
fig,axes=plt.subplots(nrows=len(years),figsize=(12,10),sharex=True)

for i,year in enumerate(years):
    data_year=Sales_by_month[Sales_by_month['Year']==year]
    axes[i].plot(data_year['Month'],data_year['Sales'],marker='o',linestyle='-',label=f'Year {year}')
    axes[i].set_title(f'Sales data for year {year}')
    axes[i].set_ylabel('sales')
    axes[i].set_xticks(range(1, 13))  # Set x-axis ticks for months
    axes[i].set_xticklabels(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
    axes[i].legend()
plt.xlabel('Month-Year')
plt.xticks(rotation=45)
plt.show()
```

```python
# yearwise sales
Yearwise_sales = data.groupby('Year')['Sales'].sum().reset_index()
Yearwise_sales
```

```python
plt.plot(Yearwise_sales['Year'],Yearwise_sales['Sales'],marker='o',linestyle='-',label=f'Year {year}')
plt.ticklabel_format(style='plain',axis='y')
plt.ticklabel_format(style='plain',axis='x')
plt.title('Year Wise Sales')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.xticks(Yearwise_sales['Year'])
plt.show()
```

```python
# Best selling products
best_selling = data.groupby('Product Name')['Sales'].sum().reset_index()
best_selling = best_selling.sort_values(by='Sales',ascending=False).head(5)
best_selling
```

```python
# Plotting the chart
plt.figure(figsize=(10, 6))
plt.bar(best_selling['Product Name'], best_selling['Sales'], color='skyblue')
plt.xlabel('Product Name')
plt.ylabel('Sales')
plt.title('Top selling products')
plt.xticks(rotation=45, ha='right')  # Rotating x-axis labels for better readability
plt.tight_layout()
plt.show()
```
